```python
jsonl_file_path = 'data_abe.jsonl'
Results_abe = []

# Abe has 3 annotations fewer than Mengyuan, so use the input has as a filter
abe_set = set()

with open(jsonl_file_path, 'r') as file:
    for line in file:
        json_obj = json.loads(line)
        Results_abe.append(json_obj["accept"][0])
        abe_set.add(json_obj["_input_hash"])
        print(json_obj)
        # break

Results_abe
```

```python
jsonl_file_path = 'data_mengyuan.jsonl'
Results_mengyuan = []
mengyuan_set = set()

with open(jsonl_file_path, 'r') as file:
    for line in file:
        json_obj = json.loads(line)
        if json_obj["_input_hash"] not in abe_set:
          print("missing:", json_obj)
        elif json_obj["_input_hash"] in mengyuan_set:
          print("duplicate:", json_obj)
        else:
          mengyuan_set.add(json_obj["_input_hash"])
          Results_mengyuan.append(json_obj["accept"][0])
        # print(json_obj)
        # break

Results_mengyuan
```

```python
mapping = {
    'POSITIVE': 1,
    'NEGATIVE': 2,
    'NEUTRAL': 3,
    'MIXED': 4,
    'UNSURE': 5
}

num_abe = [mapping[a] for a in Results_abe]
num_mengyuan = [mapping[a] for a in Results_mengyuan]
```

```python
from sklearn.metrics import cohen_kappa_score

# Calculate Cohen's Kappa
kappa = cohen_kappa_score(num_abe, num_mengyuan[:175])

print("Cohen's Kappa:", kappa)
```

```python
from sklearn.metrics import confusion_matrix,  ConfusionMatrixDisplay, classification_report
cm = confusion_matrix(Results_abe, Results_mengyuan, labels=["POSITIVE", "NEGATIVE", "NEUTRAL", "MIXED", "UNSURE"])
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["POSITIVE", "NEGATIVE", "NEUTRAL", "MIXED", "UNSURE"])
cm_display.plot()
```
